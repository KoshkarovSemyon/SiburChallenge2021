Данные взяты с сайта finam.ru

```python
import pandas as pd
import os
```

```python
data = None
files = ['CNYCB','RUBEUR','RUBUSD','USDCNY','INDUSDX','ICE.BRN']
for f in files:
    print(f)
    df = pd.read_csv(f"exdata/{f}.csv",sep=';',parse_dates=['<DATE>'])
    df['month'] = df['<DATE>'].apply(lambda x:x.replace(day=1))
    grouped = df.groupby('month')['<OPEN>']
    if data is None:
        data = pd.DataFrame([], index=grouped.first().index)
    data[f"{f}_mean"] = grouped.mean()
    data[f"{f}_mean_d"] = (data[f"{f}_mean"] - data[f"{f}_mean"].shift(1)).fillna(0)
    data[f"{f}_std"] = grouped.std()
data = data.fillna(0)
```
